```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_excel('data.xlsx')
df.head(1)
```

```python
df.Estado.value_counts().plot(kind = 'bar', color='pink', xlabel='Quantidade')
plt.title('Concentração de clientes por estado')
```

```python
df.Sexo.value_counts().reset_index
```

```python
df.Sexo.value_counts().plot(kind = 'pie')
```
